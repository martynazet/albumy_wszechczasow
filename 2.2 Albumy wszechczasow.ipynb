```python
import pandas as pd
```

```python
data = pd.read_html('https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551/', header=0)
```

```python
albums = data[0]
albums
```

```python
albums.rename(columns={'POS':'POZ', 'TITLE':'TYTUŁ', 'ARTIST':'ARTYSTA', 'YEAR':'ROK', 'HIGH POSN':'MAX POZ'}, inplace=True)
albums
```

```python
#liczba wykonawców
albums['ARTYSTA'].nunique()
```

1. Na liście znajduje się 47 wykonawców.

```python
#najczęściej na liście
albums['ARTYSTA'].value_counts()[:5]
```

2. Najcześciej na liście występują (3 razy) Coldplay oraz Take That.

```python
#zmiana nagłówków
albums.rename(str.title, axis='columns', inplace=True)
```

```python
#wyrzucenie max poz
albums.drop('Max Poz', axis=1, inplace=True)
```

```python
#najwięcej albumów - rok
albums['Rok'].value_counts()[:5]
```

6. Najwiećej albumów z listy wydano w 2000 roku.

```python
#albumy lata 1960-1990
albums_years = albums[(albums['Rok']>=1960) & (albums['Rok']<=1990)]
len(albums_years)
```

7. Na liście znajdują się 22 albumy wydane w latach 1960-1990.

```python
#najmłodszy album
albums.sort_values(by='Rok', ascending=False)[:2]
```

8. Najmłodszy album na liście wydano w 2015 roku.

```python
#najwcześniej wydane albumy - przygotowanie csv
najwczesniejsze_albumy = albums.loc[albums.groupby('Artysta')['Rok'].idxmin()]
najwczesniejsze_albumy.to_csv('najwczesniejsze_albumy.csv', index=False)
```
